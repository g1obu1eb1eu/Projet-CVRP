<h1><u>Livrable 1</u></h1>

<h2><u>Contexte</u></h2>

Ce tutoriel est à faire impérativement par les élèves. Il présente des capacités avancées de Python qui seront utilisées dans l'ensemble des Workshop (à l'exception du second), et ces fonctionnalités sont considérées comme maîtrisées par les élèves.
    
Le focus principal de ce tutoriel est sur la performance de code. Le sujet n'est pas couvert de manière exhaustive, mais se concentre sur les aspects qui concernent directement ce bloc. Une ouverture au sujet dans son ensemble est proposée à la fin (notamment autour des outils de profiling), mais il n'y a rien d'obligatoire.
    
Les élèves doivent exécuter l'ensemble des cellules, et comprendre les explications correspondantes. Idéalement, ils expérimentent à partir des exemples. Attention, certaines cellules produisent une erreur, le but est d'illustrer les limites du langage.
